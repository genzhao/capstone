```python
zappos = pd.read_csv('./zappos.csv')
six = pd.read_csv('./6pm.csv')
```

```python
zapposmerged = pd.concat(frames, axis=0, ignore_index=True)
```

```python
zapposmerged.brand_name.value_counts()
```

```python
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.lower())
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' kids',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' golf',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' kids',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' performance',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' classics',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' skateboarding',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' originalss',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' originals',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' originalss',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' originalss',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' outdoor',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' tiger',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace('onitsuka by ',''))
zappos_clean.brand_name = zappos_clean.brand_name.apply(lambda x: x.replace(' running',''))
```

```python
zap = zappos_clean
```
